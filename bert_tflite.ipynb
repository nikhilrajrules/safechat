```
import numpy as np
import os

import tensorflow as tf
assert tf.__version__.startswith('2')

from tflite_model_maker import configs
from tflite_model_maker import model_spec
from tflite_model_maker import text_classifier
from tflite_model_maker import TextClassifierDataLoader
import pandas as pd
```

```
spec = model_spec.get('mobilebert_classifier')
```

```
data_dir="/content/drive/My Drive/minor2/dataset"
train_data = TextClassifierDataLoader.from_csv(
      filename=os.path.join(os.path.join(data_dir, 'eng_train.csv')),
      text_column='sentence',
      label_column='label',
      model_spec=spec,
      delimiter=',',
      is_training=True)
test_data = TextClassifierDataLoader.from_csv(
      filename=os.path.join(os.path.join(data_dir, 'eng_test.csv')),
      text_column='sentence',
      label_column='label',
      model_spec=spec,
      delimiter=',',
      is_training=False)
```

```
model = text_classifier.create(train_data, model_spec=spec)
```

```
loss, acc = model.evaluate(test_data)
```

```
config = configs.QuantizationConfig.create_dynamic_range_quantization(optimizations=[tf.lite.Optimize.OPTIMIZE_FOR_LATENCY])
config._experimental_new_quantizer = True
```

```
model.export(export_dir='/content/drive/My Drive/minor2/mobilebert/', quantization_config=config)
```
