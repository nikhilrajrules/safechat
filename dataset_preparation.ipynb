```
import pandas as pd
```

```
df_train=pd.read_csv("/content/drive/My Drive/minor2/dataset/train.csv")
df_test=pd.read_csv("/content/drive/My Drive/minor2/dataset/test.csv")
```

```
df_train.head()
```

```
df_train=df_train.drop(['id','severe_toxic','obscene','threat','insult','identity_hate'],axis=1)
```

```
df_train.rename(columns = {'comment_text':'sentence', 'toxic':'label'}, inplace = True) 
```

```
df_test.head()
```

```
df_test_label=pd.read_csv("/content/drive/My Drive/minor2/dataset/test_labels.csv")
```

```
df_test_label.head()
```

```
df_test_label=df_test_label.drop(['severe_toxic','obscene','threat','insult','identity_hate'],axis=1)
```

```
def zero_to_one(n):
  if n==-1:
    return int(0)
  elif n==0:
    return int(1)
```

```
df_test_label['toxic']=df_test_label['toxic'].apply(zero_to_one)
```

```
df_test_label.head()
```

```
df_test=pd.merge(df_test, df_test_label,how='inner', on='id')
```

```
df_test.head()
```

```
df_test=df_test.drop(['id'],axis=1)
df_test.rename(columns = {'comment_text':'sentence', 'toxic':'label'}, inplace = True) 
```

```
df_test.head()
```

```
df_train.to_csv("/content/drive/My Drive/minor2/dataset/eng_train.csv",index=False)
df_test.to_csv("/content/drive/My Drive/minor2/dataset/eng_test.csv",index=False)
```
